Descriptive statistics, correlations, and scatter plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
df = pd.read_csv(ROOT / "data/processed/moutai_profit_processed.csv")
df[["LNY","LNX2","LNX3","LNX4","LNX5"]].describe()
```

```python
df[["LNY","LNX2","LNX3","LNX4","LNX5"]].corr()
```

```python
for col in ["LNX2","LNX3","LNX4","LNX5"]:
    plt.figure()
    plt.scatter(df[col], df["LNY"])
    plt.xlabel(col)
    plt.ylabel("LNY")
    plt.title(f"LNY vs {col}")
    plt.show()
```
