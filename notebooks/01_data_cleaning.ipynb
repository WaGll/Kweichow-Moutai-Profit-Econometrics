Load raw Moutai data, create logarithmic and differenced variables.

```python
import pandas as pd
import numpy as np
from pathlib import Path

ROOT = Path.cwd().parents[0] if Path.cwd().name == "notebooks" else Path.cwd()
df = pd.read_csv(ROOT / "data/raw/moutai_profit_2000_2023.csv")
df.head()
```

```python
rename_map = {
    "net_profit_y_billion_cny": "Y",
    "operating_revenue_x2_billion_cny": "X2",
    "operating_cost_x3_billion_cny": "X3",
    "gdp_growth_index_x4_prev_year_100": "X4",
    "cpi_x5_1978_100": "X5",
}
df = df.rename(columns=rename_map)
for col in ["Y", "X2", "X3", "X4", "X5"]:
    df[f"LN{col}"] = np.log(df[col])
    df[f"DLN{col}"] = df[f"LN{col}"].diff()
df.to_csv(ROOT / "data/processed/moutai_profit_processed.csv", index=False)
df.head()
```
